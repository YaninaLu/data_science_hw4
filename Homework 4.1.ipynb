Є набір даних speed, який представляє значення швидкості для деякого транспортного засобу в певний момент спостереження. Очевидно, що дані мають дискретний вигляд. Відомо, що спостереження відбувалися з періодом в одну годину.

speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
``

1. Приймемо, що в нульовій координаті у нас швидкість 25 км/год, відповідно до набору даних speed. Помістіть у змінну time - вектор часу, створений за допомогою np.linspace (всього 12 спостережень, від 0 до 11 годин)

4. Виконайте інтерполяцію за допомогою interp1d(kind='cubic') та отримайте функцію на 10000 значень. Побудуйте безперервний графік отриманої функції.
5. Обчисліть інтеграл для отриманої інтерполяційної функції на проміжку [0, 11]

Виконайте також пункти 4 та 5 для kind='quadratic'.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.interpolate import interp1d
from scipy import integrate
```

```python
time = np.linspace(0, 11, num=12)
print(time)
```

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
```

```python
plt.figure(figsize=(10, 8))
sns.set_style('whitegrid')
sns.scatterplot(x=time, y=speed).set(xlim=(0, 11), ylim=(0, 130))
plt.title('Speed vs Time')
```

```python
func = interp1d(time, speed, kind='cubic')
```

```python
points = np.linspace(0, 11, 1000)
plt.figure(figsize=(10, 8))
sns.set_style('whitegrid')
sns.scatterplot(x=time, y=speed).set(xlim=(0, 11), ylim=(0, 130))
plt.title('Speed vs Time Interpolated Cubic')
sns.lineplot(x=points, y=func(points), color='red', linestyle='--')
```

```python
res = integrate.quad(func, 0, 11)
print(res)
```

Quadratic interpolation:

```python
quadr_func = interp1d(time, speed, kind='quadratic')
plt.figure(figsize=(10, 8))
sns.set_style('whitegrid')
sns.scatterplot(x=time, y=speed).set(xlim=(0, 11), ylim=(0, 130))
plt.title('Speed vs Time Interpolated Quadratic')
sns.lineplot(x=points, y=quadr_func(points), color='green')
```

```python
res = integrate.quad(quadr_func, 0, 11)
print(res)
```
